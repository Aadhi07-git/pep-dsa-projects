This notebook performs sentiment analysis on cricket-related tweets using Python.
It includes data exploration, sentiment labeling, and visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("data/cricket_tweets.csv")

# Check data
print(df.shape)
df.head()
```

```python
def label_sentiment(text):
    text = str(text).lower()
    if any(word in text for word in ["win", "great", "love", "best", "amazing"]):
        return "positive"
    elif any(word in text for word in ["bad", "worst", "hate", "lose", "poor"]):
        return "negative"
    else:
        return "neutral"

df["sentiment"] = df["text"].apply(label_sentiment)
df["sentiment"].value_counts()
```

```python
sentiment_counts = df["sentiment"].value_counts()

plt.figure()
sentiment_counts.plot(kind="bar")
plt.title("Sentiment Distribution of Cricket Tweets")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")

plt.savefig("sentiment_bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```
